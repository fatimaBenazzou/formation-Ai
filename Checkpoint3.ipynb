3. Split your dataset into a training set and a testing set.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
```

```python
train_data,test_data=train_test_split(df,train_size=0.8,random_state=3)
```

```python
X_train=np.array(train_data['sqft_living']).reshape(-1,1)
y_train=np.array(train_data['price']).reshape(-1,1)

X_test=np.array(test_data['sqft_living']).reshape(-1,1)
y_test=np.array(test_data['price']).reshape(-1,1)
```

4. Apply Linear regression to your training set.

```python
# Initialisation du modÃ¨le
model = LinearRegression()

# EntraÃ®nement du modÃ¨le sur l'ensemble d'entraÃ®nement
model.fit(X_train, y_train)
```

5. Plot the linear regression.

```python
_, ax = plt.subplots(figsize= (12, 10))
plt.scatter(X_test, y_test, color= 'pink', label = 'data')
plt.plot(X_test, model.predict(X_test), color='red', label= ' Predicted Regression line')
plt.xlabel('Living Space (sqft)')
plt.ylabel('price')
plt.legend()
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
```

6. Measure the performance of linear regression using the testing set.

```python
# PrÃ©dictions sur l'ensemble de test
y_pred = model.predict(X_test)
print('linear model')

# Calcul des mÃ©triques de performance
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r_squared = model.score(X_test, y_test)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error: {round(mae, 2)}') 
print(f'Mean Squared Error: {round(mse, 2)}') 
print(f'R Mean Squared Error: {round(rmse, 2)}') 
print(f'R-squared: {round(r_squared, 3)}')  
```

7. Apply multi-linear regression and compare it to the linear model.

```python
fig,ax=plt.subplots(2,1,figsize=(15,10))
sns.boxplot(x=train_data['grade'],y=train_data['price'],ax=ax[0])
sns.boxplot(x=train_data['bedrooms'],y=train_data['price'],ax=ax[1])
_ , axes = plt.subplots(1, 1, figsize=(15,10))
sns.boxplot(x=train_data['bathrooms'],y=train_data['price'])
```

```python
features1=['bedrooms','grade','sqft_living','sqft_above']

modelMult=LinearRegression()
modelMult.fit(train_data[features1],train_data['price'])
pred=modelMult.predict(test_data[features1])
print('complex_model 1')

mse_multi=metrics.mean_squared_error(y_test,pred)
print('R mean squared error(RMSE)', round((mse_multi),2))
print('R squared',round(modelMult.score(train_data[features1],train_data['price']),3))
```

```python
# Comparaison des erreurs
if mse_multi < mse:
    print("La rÃ©gression multi a une erreur quadratique moyenne infÃ©rieure, elle semble Ãªtre plus performante.")
else:
    print("La rÃ©gression linÃ©aire a une erreur quadratique moyenne infÃ©rieure, elle semble Ãªtre plus performante.")
```

8. Apply Polynomial regression and compare it to linear and ultilinear regression.

```python
# Assuming X_train and X_test are your feature matrices, and y_train, y_test are your target variables

polyfeat=PolynomialFeatures(degree=2)
xtrain_poly=polyfeat.fit_transform(train_data[features1])
xtest_poly=polyfeat.fit_transform(test_data[features1])

poly=LinearRegression()
poly.fit(xtrain_poly,train_data['price'])
polypred=poly.predict(xtest_poly)

print('Complex Model_3')
mse_poly = metrics.mean_squared_error(test_data['price'], polypred)
print('R Mean Squared Error (RMSE) ', round((mse_poly), 2))
print('R-squared ', round(poly.score(xtrain_poly, train_data['price']), 3))
```

```python
# Comparaison des modÃ¨les
if mse_multi < mse and mse_multi < mse_poly:
    print("La rÃ©gression multi a une erreur quadratique moyenne infÃ©rieure, elle semble Ãªtre plus performante.")
elif mse_poly < mse:
    print("La rÃ©gression polynomiale a une erreur quadratique moyenne infÃ©rieure, elle semble Ãªtre plus performante.")
else:
    print("La rÃ©gression linÃ©aire a une erreur quadratique moyenne infÃ©rieure, elle semble Ãªtre plus performante.")
```
