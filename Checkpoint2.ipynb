```python
def plot_correlation_map(df):
    corr = df.corr()
    s, ax = plt.subplots(figsize=(12, 10))
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    s = sns.heatmap(
        corr,
        cmap=cmap,
        square=True,
        cbar_kws={'shrink': .9},
        ax=ax,
        annot=True,
        annot_kws={'fontsize': 12}
    )

plot_correlation_map(df_encoded)

# The provided function creates a heatmap showing the correlation between numerical variables. 
```

Analysis:

The function plot_correlation_map is a utility for visualizing correlations between numerical variables in a dataset. The resulting heatmap provides an intuitive and informative way to quickly identify relationships between variables. The color intensity and the annotated correlation coefficients allow for easy interpretation. For instance, dark blue cells indicate strong negative correlations, while dark red cells indicate strong positive correlations. This visualization is particularly useful in exploratory data analysis, feature selection, and identifying multicollinearity in regression models.
